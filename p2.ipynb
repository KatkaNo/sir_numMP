```python
import numpy as np
import matplotlib.pyplot as plt

H = []
h = 0.1
Eps = []

while(h>0.001):
  H.append(h)
  print(h)
  h=h/2

for h in H:
# h = 0.1
  tau = h*(1/7)*0.95 #tau/h = a*(1/(t+6)), t = 1, a = 0.95
  sig = tau/h
  L = int(1/h)
  N = int(1/tau)

  u = np.zeros([L+1,N+1])
  u_an = np.zeros([L+1,N+1])
  u_a = np.zeros(L+1)
  u_c = np.zeros(L+1)
  x = np.linspace(0,1,L+1)
  t = np.linspace(0,1,N+1)

  for l in range(L+1):
    u[l][0] = x[l]
  for n in range(N+1):
    u[0][n] = -(t[n]**2)/2-6*t[n]
    u[1][n] = -(t[n]**2)/2-6*t[n] + h

  for l in range(2,L+1):
    for n in range(N):
      u[l][n+1] = u[l][n]+(tau/2+t[n]+6)*(sig/2)*(-u[l-2][n]+4*u[l-1][n]-3*u[l][n])+(t[n]+6)**2*((sig**2)/2)*(u[l-2][n]-2*u[l-1][n]+u[l][n])

  for l in range(L+1):
    for n in range(N+1):
        u_an[l][n] = -(t[n]**2)/2-6*t[n] + x[l]

  for l in range(L+1):
    u_c[l] = u[l][N]
    u_a[l] = -(t[N]**2)/2-6*t[N] + x[l]
    eps = np.linalg.norm(u_c[l]-u_a[l])

  Eps.append(eps)
  print(eps)

  plt.plot(x, u_c, label= "h")

plt.title("стремление к аналитическому")
plt.plot(x,u_a, label= "аналитика")
plt.legend()
plt.show()
plt.title("Аналитика")
plt.plot(x,u_a)
plt.show()
plt.loglog(H,Eps)
plt.xlabel("ln(h)")
plt.ylabel("ln(eps)")
plt.show()
```
